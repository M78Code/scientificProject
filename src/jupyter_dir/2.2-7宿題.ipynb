```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt
```

```python
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False


PI = np.pi
x = np.linspace(-PI, PI, 1000)

 
#　原始関数 y = x^2
x_original = pow(x, 2)

# フーリエ級数の第ｎ項
def f_(n):
    return (-2*PI) * (pow(-1, n) / n) * np.cos(n*x)

# フーリエ級数の値を計算する
def fourier_series_approximation(t, N):
    x_t = np.zeros_like(t)
    for n in range(1, N + 1):
        x_t += f_(n)
    return x_t

# 異なる値ｎの波形
N_values = [1,3,5]
x_approximations = [fourier_series_approximation(x, N) for N in N_values]


# 熱方程式関数
# def u(x, n):
#     return f_(n) * np.cos(n*x) * pow(sym.exp, -n)
# u_approximations = [u(t, N) for N in N_values]

# 波形の図を描く
def draw_():
    
    plt.figure(figsize=(12, 8))
 
    plt.subplot(2, 2, 1)
    plt.plot(x, x_original)
    plt.title('原始関数')
    plt.xlabel('x')
    plt.ylabel('f(x)')
 
    for i, N in enumerate(N_values):
        plt.subplot(2, 2, i + 2)
        plt.plot(x, x_approximations[i], label=f'N = {N}')
        plt.title(f'フーリエ級数近似（n = {N}）')
        plt.xlabel('時間')
        plt.ylabel('振幅')
 
    plt.tight_layout()
    plt.show()
```

```python
draw_()
```

```python
def draw_f(N):
    
    plt.figure(figsize=(12, 8))
    # plt.subplot(2, 2)
    plt.plot(x, fourier_series_approximation(x, N), label=f'N = {N}')
    plt.title(f'フーリエ級数近似（n = {N}）')
    plt.xlabel('時間')
    plt.ylabel('振幅')
 
    plt.tight_layout()
    plt.show()
```
